# Data Catagories:
 Roll No: Student number out of 10000

 Gender: Male or Female

 Race Ethicity: Not enough data

 Parental Level Of Education: Parents level of education (Bachelors, Master, Some College, High School, Some High School)

 Lunch: 0=Standard, 1=Free/Reducded Lunch

 Test Preparation Course: 0=None, 1=Completed

 Math Score: Test score out of 100

 Reading Score: Test score out of 100

 Writing Score: Test score out of 100

 Science Score: Test score out of 100

 Total Score Grade: All test scores combined out of 400

```python
# Scatterplot of math score vs reading score
df_sorted = df.sort_values(by=['math_score', 'reading_score'])
sns.scatterplot(x='math_score', y='reading_score', data=df_sorted, color='blue', label='Reading Score')
sns.scatterplot(x='reading_score', y='math_score', data=df_sorted, color='red', label='Math Score')
sns.scatterplot(x='math_score', y='reading_score', data=df_sorted)
plt.title('Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.xticks(range(0, 100, 10))
plt.yticks(range(0, 100, 10))
plt.ylim(0, 100)
plt.show()
```
